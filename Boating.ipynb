```python
# read the data
import pandas as pd
df = pd.read_csv('BoatingZones.csv')

import geojson
with open('BZ.geojson', 'r', encoding='UTF-8') as f:
    BZ = geojson.load(f)
```

```python
# filter the CSV
df = df[df['ZONETYPE']!='Activities Prohibited']
df = df[df['ZONETYPE']!='Exclusive Use / Special Purpose']
df = df[df['ZONETYPE']!='Speed Exclusion']
df = df[df['ZONETYPE']!='Vessels Prohibited']
df = df.dropna(subset=['ZONETYPE'])
df = df.reset_index()
df = df.drop('index', 1)
```

```python
# add one coordinate in the zone to data
df['longitude'] = 0
df['latitude'] = 0
for i in range(len(df)):
    df['longitude'].iloc[i] = BZ['features'][i]['geometry']['coordinates'][0][0][0][0]
    df['latitude'].iloc[i] = BZ['features'][i]['geometry']['coordinates'][0][0][0][1]
```

```python
import pandas as pd
BZ = pd.read_csv('BZ.csv')
```

```python
# clean the data
ID = list(BZ['UNIQUEID'])

for i in range(len(ID)):
    ID[i] = ID[i][:-5]
    ID[i] = ID[i].replace('-',' ')
    ID[i] = ID[i].replace('_','')
```

```python
BZ['UNIQUEID'] = ID
```

```python
BZ.to_csv('BZ_new.csv', index=False)
```
